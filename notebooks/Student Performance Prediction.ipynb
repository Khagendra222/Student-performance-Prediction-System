# Phase 2: Data Preprocessing

```python
# Cheack Data Types
df.info()
```

```python
# Encode Categorical Variables

from sklearn.preprocessing import LabelEncoder
import joblib

# Dictionary to store encoders
encoders = {}

# Encode categorical columns and save each encoder
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le

df.head()
```

```python
import os
import joblib

# Create models folder if it doesn't exist
os.makedirs("models", exist_ok=True)

# Save all label encoders
joblib.dump(encoders, "models/label_encoders.pkl")

print("Label encoders saved successfully!")
```

```python
# Verify Encoding
df.info()
```

```python
# Feature Selection
# We don't use:
# G3(because it directly gives the final grade)
# Result( because it is our target)

x=df.drop(['G3','Result'],axis=1)
y=df['Result']
```

```python
#Check Feature and Target Shapes

print(x.shape)
print(y.shape)
```

```python
# Train-Test Split

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
# Verify Split

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Phase 4 : Model Training 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import(
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
print('libraries  loaded')
```

```python
# Create the models

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Support Vector Machine": SVC(random_state=42)
}
```

```python
# Train All Models

for name,model in models.items():
    model.fit(x_train,y_train)

print('All models trained successfully')
```

```python
# Evaluate Models

results=[]
for name, model in models.items():
    y_pred= model.predict(x_test)

    accuracy=accuracy_score(y_test,y_pred)
    precision=precision_score(y_test,y_pred)
    recall=recall_score(y_test,y_pred)
    f1=f1_score(y_test,y_pred)

    results.append({
        'Model':name,
        'Accuracy':accuracy,
        'Precision':precision,
        'Recall':recall,
        'F1 Score':f1
    })
```

```python
# Comparision Table

import pandas as pd
results_df=pd.DataFrame(results)

results_df.sort_values(
    by='Accuracy',
    ascending=False
)
```

# Observation
Observation and Interpretation: Four machine learning algorithms were trained and evaluated to predict student performance. Logistic Regression achieved the highest accuracy of 94.94%, along with the best Precision (98.00%) and F1 Score (96.08%). Based on these evaluation metrics, Logistic Regression outperformed the other models and was selected as the final prediction model for this project.

```python
# Plot Model Accuracy

import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(results_df['Model'],results_df['Accuracy'])

plt.title('Model Accuracy comparision')
plt.xlabel('Machine Learning  Models')
plt.ylabel('Accuracy')
```

```python
# Confusion Matrix for  the Best Model


from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

best_model = models["Logistic Regression"]

y_pred = best_model.predict(x_test)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()
```

# Observation
Observation and Interpretation: The Logistic Regression model correctly classified 26 failing students and 49 passing students. It incorrectly classified 1 failing student as passing and 3 passing students as failing. Overall, the model made 75 correct predictions out of 79 test samples, achieving an accuracy of 94.94%. The low number of misclassifications indicates that the model is highly effective in predicting student academic performance.

 # Phase 5: Model Optimization & Saving

```python
# Hyperparameter Tuning(Logistic Regression)

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define parameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs'],
    'max_iter': [1000]
}

# Create GridSearchCV object
grid_search = GridSearchCV(
    estimator=LogisticRegression(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Train
grid_search.fit(x_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Cross Validation Accuracy:", round(grid_search.best_score_, 4))
```

```python
# Use the Best Model
best_model=grid_search.best_estimator_
print(best_model)
```

```python
# Evaluate the Tuned Model

from sklearn.metrics import accuracy_score
y_pred=best_model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f'Test Accuracy:{accuracy:.4f}')
```

```python
# Classification Report

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title("Confusion Matrix - Tuned Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

# Observation
The confusion matrix illustrates that the majority of student records were classified correctly. Only a small number of students were misclassified, confirming that the optimized Logistic Regression model provides reliable predictions for student performance.

```python
# Feature Importance ( Logistic Regression)

import pandas as pd

feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': best_model.coef_[0]
})

feature_importance['Absolute Coefficient'] = feature_importance['Coefficient'].abs()

feature_importance = feature_importance.sort_values(
    by='Absolute Coefficient',
    ascending=False
)

feature_importance.head(10)
```

# Observation
The feature importance analysis identifies the variables that have the greatest influence on predicting student performance. Features with larger absolute coefficients contribute more significantly to the model's decision-making process.

```python
# Visualize Feature Importance\

import matplotlib.pyplot as plt

# Sort by absolute coefficient
top_features = feature_importance.sort_values(
    by='Absolute Coefficient',
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

bars = plt.barh(
    top_features['Feature'],
    top_features['Absolute Coefficient']
)

plt.gca().invert_yaxis()

plt.title("Top 10 Most Important Features", fontsize=16, fontweight='bold')
plt.xlabel("Absolute Coefficient", fontsize=12)
plt.ylabel("Features", fontsize=12)

# Add coefficient values on each bar
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 0.02,
        bar.get_y() + bar.get_height()/2,
        f"{width:.2f}",
        va='center',
        fontsize=10
    )

plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

# Observation
The feature importance analysis reveals that G2 (second-period grade) is the most influential feature with an absolute coefficient of 1.15, followed by G1 (first-period grade) with 0.42. Other important features include age (0.41), family relationship (famrel) (0.29), study time (0.23), activities (0.21), weekend alcohol consumption (Walc) (0.21), father's education (Fedu) (0.21), mother's job (Mjob) (0.18), and parental status (Pstatus) (0.14). These results indicate that previous academic performance has the strongest influence on predicting a student's final performance, while demographic, family, and lifestyle factors also contribute to the prediction.
